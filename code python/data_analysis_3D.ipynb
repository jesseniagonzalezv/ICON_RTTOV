```python
#-- input file rectilinear_grid_3d.nc from the NCL User Guide
#-- is available in the PyNGL installation
fname_coarse = '/poorgafile1/climate/hdcp2/2013/3d_coarse_day_ll_DOM01_ML_20130502T180000Z.nc'
fname2_coarse = '/poorgafile1/climate/hdcp2/2013/3d_coarse_day_ll_DOM02_ML_20130502T180000Z.nc'
fname3_coarse = '/poorgafile1/climate/hdcp2/2013/3d_coarse_day_ll_DOM03_ML_20130502T180000Z.nc'


fname = '/poorgafile1/climate/hdcp2/2013/3d_fuzzy_hope_ll_DOM01_ML_20130502T180000Z.nc'
fname2 = '/poorgafile1/climate/hdcp2/2013/3d_fuzzy_hope_ll_DOM02_ML_20130502T180000Z.nc'
fname3 = '/poorgafile1/climate/hdcp2/2013/3d_fuzzy_hope_ll_DOM03_ML_20130502T180000Z.nc'

#-- open file
ds = xr.open_dataset(fname)
ds2= xr.open_dataset(fname2)
ds3= xr.open_dataset(fname3)

ds_c = xr.open_dataset(fname_coarse)
ds2_c= xr.open_dataset(fname2_coarse)
ds3_c= xr.open_dataset(fname3_coarse)

ds.values
```

## Variables to use

- Temperature ta(time, height, lat, lon) 
		ta:standard_name = "air_temperature" ;
		ta:long_name = "Temperature" ;
		ta:units = "K" ;
		ta:param = "0.0.0" ;
        
- qv? hus (time, height, lat, lon)2.01304e+07 max value 2.01305e+07 min value
		hus:standard_name = "specific_humidity" ;
		hus:long_name = "Specific humidity" ;
		hus:units = "kg kg-1" ;
		hus:param = "0.1.0" ;
        
- ql? clw (time, height, lat, lon)2.01304e+07 max value 2.01305e+07
		clw:standard_name = "qc" ;
		clw:long_name = "specific_cloud_water_content" ;
		clw:units = "kg kg-1" ;
		clw:param = "22.1.0" ;
        
- Nd/qnc (time, height, lat, lon)
		qnc:standard_name = "qnc" ;
		qnc:long_name = "number_concentration_cloud_droplets" ;
		qnc:units = " kg-1" ;
		qnc:param = "28.6.0"
        
- pres (time, height, lat, lon)
		pres:standard_name = "air_pressure" ;
		pres:long_name = "Pressure" ;
		pres:units = "Pa" ;
		pres:param = "0.3.0" ;

### export ta,pres, hus in a CSV

```python
import csv
myData = [["Temperature/ta(K)", "air_pressure/pres(Pa)", "specific_humidity/hus(kg kg-1)"]]              
myFile = open('profiles.csv', 'w')
with myFile:
    writer = csv.writer(myFile)
    writer.writerows(myData)  #list
    
for n in range(len(ta)):
    myData = [[ta[n],pres[n],hus[n]]] 
    myFile = open('profiles.csv', 'a')
    with myFile:
        writer = csv.writer(myFile)
        writer.writerows(myData)  #list
```
